```python
def VMDV(f):
    f=f
    alpha = f.shape[0]*1.5
    u, u_hat, omega = VMD(f.values, alpha, tau, K, DC, init, tol)
    plt.figure()
    plt.plot(u.T)
    plt.title('Decomposed modes')
    fig1 = plt.figure()
    plt.plot(f.values)
    fig1.suptitle('Original input signal and its components')
    fig = plt.figure(figsize=(20,10))

    for i in range(1,K+1):
        a = u[i-1,:]
        ax=fig.add_subplot(K,1,i)
        ax.plot(a)
        ax.set_ylabel("IMF"+str(i))

    plt.tight_layout()
    plt.show()


def VMDValue(f):
 f = f
 alpha = f.shape[0] * 1.5
 u, u_hat, omega = VMD(f.values, alpha, tau, K, DC, init, tol)
 return u
#. some sample parameters for VMD
  # moderate bandwidth constraint
tau = 0.  # noise-tolerance (no strict fidelity enforcement)
K = 5  # 3 modes
DC = 1  # no DC part imposed
init = 1  # initialize omegas uniformly
tol = 1e-7
```

```python
def runticc(savename,clustersavename,data,clusterN):
    np.savetxt(savename,data,fmt='%.3f', delimiter=',')
    fname = savename
    ticc= TICC(window_size=1, number_of_clusters=clusterN, lambda_parameter=11e-2, beta=600, maxIters=100, threshold=2e-5,
    write_out_file=False, prefix_string="output_folder/", num_proc=1,compute_BIC=True)
    (cluster_assignment, cluster_MRFs,TICCbic) = ticc.fit(input_file=fname)
    np.savetxt(clustersavename, cluster_assignment, fmt='%d', delimiter=',')
    return ticc,TICCbic
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
```

```python
def ploty(data,bic, name,ax):

    silhouettescore=[]
    calinskiscore=[]
    daviesscore=[]
    for j,i in enumerate(range(2,7)):
        clustersavename="result_cluster"+str(i)+name
        cluster=pd.read_table(clustersavename, sep=',', header=None)
        temp=data.copy()
        temp["cluster"]=cluster.values
        score1=silhouette_score(temp.drop(columns="cluster"),cluster)
        score2=calinski_harabasz_score(temp.drop(columns="cluster"),cluster)
        score3=davies_bouldin_score(temp.drop(columns="cluster"),cluster)
        silhouettescore.append(score1)
        calinskiscore.append(score2)
        daviesscore.append(score3)



    # fig, ax = plt.subplots(figsize=(15,5))
    # fig.subplots_adjust(right=0.75)

    twin1 = ax.twinx()
    twin2 = ax.twinx()
    twin3 = ax.twinx()

    # Offset the right spine of twin2.  The ticks and label have already been
    # placed on the right by twinx above.
    twin2.spines.right.set_position(("axes", 1.08))
    twin3.spines.right.set_position(("axes", 1.16))

    p1, = ax.plot(range(2, 7), bic, "b-", label="BIC")
    p2, = twin1.plot(range(2, 7), silhouettescore, "r-", label="Silhouette Score")
    p3, = twin2.plot(range(2, 7), calinskiscore, "g-", label="Calinski Harabasz Score")
    p4, = twin3.plot(range(2, 7), daviesscore, "y-", label="Davies Bouldin Score")

    ax.set_xlim(1, 6)
    # ax.set_ylim(0, 2)
    # twin1.set_ylim(0, 4)
    # twin2.set_ylim(1, 65)

    ax.set_xlabel("Cluster Number")
    ax.set_ylabel("BIC")
    twin1.set_ylabel("Silhouette Score")
    twin2.set_ylabel("Calinski Harabasz Score")
    twin3.set_ylabel("Davies Bouldin Score")

    ax.yaxis.label.set_color(p1.get_color())
    twin1.yaxis.label.set_color(p2.get_color())
    twin2.yaxis.label.set_color(p3.get_color())
    twin3.yaxis.label.set_color(p4.get_color())

    tkw = dict(size=4, width=1.5)
    ax.tick_params(axis='y', colors=p1.get_color(), **tkw)
    twin1.tick_params(axis='y', colors=p2.get_color(), **tkw)
    twin2.tick_params(axis='y', colors=p3.get_color(), **tkw)
    twin3.tick_params(axis='y', colors=p4.get_color(), **tkw)
    ax.tick_params(axis='x', **tkw)

    ax.legend(handles=[p1, p2, p3, p4])

    # plt.show()
```

```python
fig, axs = plt.subplots(3,1,figsize=(15, 15))
ax=plt.subplot(3,1,1)
ploty(data1,bicAug,"TICCCLUSTER_Aug.txt",ax)
ax.set_title("Aug. TICC Cluster")
# plt.axvline(4)

ax=plt.subplot(3,1,2)
ploty(data2,bicSept,"TICCCLUSTER_Sept.txt",ax)
ax.set_title("Sep. TICC Cluster")
# plt.axvline(4)

ax=plt.subplot(3,1,3)
ploty(data3,bicOct,"TICCCLUSTER_Oct.txt",ax)
ax.set_title("Oct. TICC Cluster")
# plt.axvline(3)

fig.tight_layout()
plt.savefig("TICC CLUSTERNUMBER.svg")
plt.show()
```
